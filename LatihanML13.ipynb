PENJUMLAHAN VEKTOR

```
import matplotlib.pyplot as plt
v = np.array([3,0])
w = np.array([-3,-3])
print(v)
print(w)
s = v + w
print(s)

#tentukan vektor origin
origin = [0],[0]
vecs = np.array([v,w,s])

#plot
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['r','b', 'g'], scale=10)
plt.show()
```

dengan

dot product of $\vec{v}\cdot \vec{w}$

```
v = np.array([3,2])
w = np.array([10, 5])

Dot1= np.dot(v,w)
print(Dot1)

Dot2= v @ w
print(Dot2)
```

Cosine Rule $\vec{v}\cdot \vec{w}= ||\vec{v}|| ||\vec{w}|| \cos(\theta)$

```
CosThet= (v@w)/(np.linalg.norm(v)*np.linalg.norm(w))
print(CosThet)

theta = math.acos(CosThet)
print(theta)

thetaDeg= math.degrees(theta)
print(thetaDeg)
```

CROSS Product Multiplication

```
v1 = np.array([1, 2, 3])
v2 = np.array([3, 2, 1])

Cross = np.cross(v1,v2)
print(Cross)
```

# FINITE DIFFERENCE METHOD 

Untuk menghampiri turunan suatu fungsi, menggunakan teori limit yakni

 $f'(x) = \lim_{h\rightarrow 0} \frac{f(x+h) - f(x)}{h}$
 
 Jadi, secara numerik jika domain adalah $[0, L]$ dan domain dibagi menjadi N+1 titik, maka $h = \frac{L}{N}$.
 
 Diberikan data berikut ini,

1.   Nilai $x_i=h\times i,\forall i \in \{0,1, \cdots N\}$
2.   Inner index didefinisikan pada $i\in \{1,2, \cdots N\}$
3.   Index Batas didefinisikan pada $i \in \{0,N+1\}$


Hampiran turunan suatu fungsi $f(x)$ menggunakan **Finite Difference Method --> Forward difference** adalah



1.   Untuk setiap titik di Inner index dan index $i=0$,

      $df(x_i) \approx \displaystyle\frac{f(x_{i+1}) - f(x_i)}{h}$

2.   Sedangkan untuk batasnya yakni di titik $N+1$ gunakan Backward,
     
     $df(x_{N+1}) \approx \displaystyle \frac{f(x_{N}) - f(x_{N+1})}{h}$


Hampiran turunan suatu fungsi $f(x)$ menggunakan **Finite Difference Method --> Backward difference** adalah



1.   Untuk setiap titik di Inner index dan index $i=N+1$,

      $df(x_i) \approx \displaystyle\frac{f(x_{i}) - f(x_{i-1})}{h}$

2.   Sedangkan untuk batasnya yakni di titik $0$ gunakan Backward,
     
     $df(x_0) \approx \displaystyle \frac{f(x_{1}) - f(x_{0})}{h}$

 
Hampiran turunan suatu fungsi $f(x)$ menggunakan **Finite Difference Method --> Central difference** adalah



1.   Untuk setiap titik di Inner index,

      $df(x_i) \approx \displaystyle\frac{f(x_{i+1}) - f(x_{i-1})}{2h}$

2.   Sedangkan untuk batasnya yakni di titik $0$ dan $N+1$ gunakan Backward dan Forward,
     
     $df(x_0) \approx \displaystyle \frac{f(x_{1}) - f(x_{0})}{h}$
     
     $df(x_{N+1}) \approx \displaystyle \frac{f(x_{N}) - f(x_{N+1})}{h}$
