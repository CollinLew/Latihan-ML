Buatlah atau plot fungsi \\

untuk $\forall x \in [0, 10]$


1.   $f(x)=c$
2.   $f(x)=2x+3$
3.   $f(x)=\frac{2}{100} x + e^{-8}$



```
import numpy as np
from matplotlib import pyplot as plt
import math

def f1(x):
  return x**2 +1

def f2(x):
  return 2*x - 100
def f3(x):
  return (2/100)*x + math.exp(-8)
def f4(x):
  return x**2 + 1

x = np.array(range(0, 11))
print(x)
nx = 21
x1 = np.linspace(0,10,nx)
print(x1[20])

plt.plot(x1,f2(x1))
plt.plot(x1,f4(x1))
plt.show()
```

# MENGHITUNG GRADIENT

```
X_1= x1[6]
X_2 = x1[10]

DX= X_2 -X_1

#Gradient untuk fungsi f2
DF = f2(X_2) - f2(X_1)
m=DF/DX
print(m)
#Gradient untuk fungsi f3
DF = f3(X_2) - f3(X_1)
m=DF/DX
print(m)
```
